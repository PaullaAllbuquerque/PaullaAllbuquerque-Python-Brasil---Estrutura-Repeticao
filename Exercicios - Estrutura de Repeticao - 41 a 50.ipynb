```python
'''Exercício 41 - Faça um programa que receba o valor de uma dívida e mostre uma tabela com os seguintes dados: 
valor da dívida, valor dos juros, quantidade de parcelas e valor da parcela.
Os juros e a quantidade de parcelas seguem a tabela abaixo:
Quantidade de Parcelas  % de Juros sobre o valor inicial da dívida
1       0
3       10
6       15
9       20
12      25
Exemplo de saída do programa:
Valor da Dívida Valor dos Juros Quantidade de Parcelas  Valor da Parcela
R$ 1.000,00     0               1                       R$  1.000,00
R$ 1.100,00     100             3                       R$    366,00
R$ 1.150,00     150             6                       R$    191,67'''

parcelas=[1,3,6,9,12]
juros=[0,10,15,20,25]

div=int(input('Insira o valor da dívida: '))
print('{:<10}{:>18}{:>26}{:>20}'.format('Valor da dívida','Valor do juros','Quantidade de parcelas','Valor da parcela'))
print('{:<10}{:>18}{:>26}{:>20}'.format('-'*16,'-'*16,'-'*23,'-'*18))

for d in range(0,len(parcelas)):
    vJuros=((juros[d]/100)*div)
    vDivida=div+vJuros
    vParcela=vDivida/parcelas[d]
    print('{:<10}{:>18}{:>26}{:>20}'.format(vDivida,round(vJuros),parcelas[d],round(vParcela,2)))
```

```python
'''Exercício 42 - Faça um programa que leia uma quantidade indeterminada de números positivos e conte quantos deles
estão nos seguintes intervalos: [0-25], [26-50], [51-75] e [76-100]. A entrada de dados deverá terminar quando for 
lido um número negativo.
'''
int1=[]
int2=[]
int3=[]
int4=[]

num=True
while num>=0:
    num=int(input('Insira um número positivo: '))
    if num>=0 and num<=25:
        int1.append(num)
    elif num>=26 and num<=50:
        int2.append(num)
    elif num>=51 and num<=75:
        int3.append(num)
    elif num>=76 and num<=100:
        int4.append(num)
    else:
        print('[NÚMERO INVÁLIDO]')
        num=int(input('Insira um número positivo: '))

        
print('\nNo intervalo entre [0-25] foram digitados: ',len(int1) ,'números')
print('No intervalo entre [26-50] foram digitados: ',len(int2) ,'números')
print('No intervalo entre [51-75] foram digitados: ',len(int3) ,'números')
print('No intervalo entre [76-100] foram digitados: ',len(int4) ,'números')
```

```python
'''Exercício 43 - O cardápio de uma lanchonete é o seguinte:
Especificação   Código  Preço
Cachorro Quente 100     R$ 1,20
Bauru Simples   101     R$ 1,30
Bauru com ovo   102     R$ 1,50
Hambúrguer      103     R$ 1,20
Cheeseburguer   104     R$ 1,30
Refrigerante    105     R$ 1,00
Faça um programa que leia o código dos itens pedidos e as quantidades desejadas. 
Calcule e mostre o valor a ser pago por item (preço * quantidade) e o total geral do pedido. 
Considere que o cliente deve informar quando o pedido deve ser encerrado.'''

comidas=['Cachorro Quente', 'Bauru Simples', 'Bauru com Ovo', 'Hambúrguer','Cheeseburguer', 'Refrigerante']
codigos=[100, 101, 102, 103,104, 105]
precos=[1.20, 1.30, 1.50, 1.20,1.30, 1.00]
lista=[]
listaGeral=[]
soma=[]

perg=True
while perg != 'N':
    pedido=int(input('Insira o código do produto: '))
    if pedido not in codigos:
        print('[CÓDIGO INVÁLIDO]')
        pedido=int(input('Insira o código do produto: '))
    else:
        quant=int(input('Insira a quantidade: '))
        for p in range(1,len(codigos)):
            if pedido==codigos[p]:
                valor=quant*precos[p]
                lista.append(codigos[p])
                lista.append(comidas[p])
                lista.append(quant)
                lista.append(valor)
                soma.append(valor)
    listaGeral.append(lista[:])
    lista.clear()
    perg=str(input('Mais algum pedido senhor? [S/N]')).upper()
print('\nPEDIDO')
print('='*30)
print('{:<5}{:>7}{:>16}{:>10}'.format('Cod.','Ítem','Quant.','Valor'))
for g in listaGeral:
    print('{:<5}{:>7}{:>5}{:>12}'.format(g[0],g[1],g[2],g[3]))
print('-'*30)
print('Total - R$', sum(soma))

```

```python
'''Exercício 44 - Em uma eleição presidencial existem quatro candidatos. Os votos são informados por meio de código. 
Os códigos utilizados são:
1 , 2, 3, 4  - Votos para os respectivos candidatos 
(você deve montar a tabela ex: 1 - Jose/ 2- João/etc)
5 - Voto Nulo
6 - Voto em Branco
Faça um programa que calcule e mostre:
O total de votos para cada candidato;
O total de votos nulos;
O total de votos em branco;
A percentagem de votos nulos sobre o total de votos;
A percentagem de votos em branco sobre o total de votos. Para finalizar o conjunto de votos tem-se o valor zero.'''

codigos=[1,2,3,4,5,6]
candidatos=['Estrôncio', 'Filomena', 'Jozenildo', 'Margareth', 'Nulo', 'Branco']
votos=[]

voto=True
nVoto=1
while voto!=0:
    voto=int(input('Insira o código do candidato: '))
    if voto==0:
        break
    else:
        while voto not in codigos:
            print('[CÓDIGO INEXISTENTE]')
            voto=int(input('Insira o código do candidato: '))
        votos.append(voto)
    nVoto+=1

cont=0
for c in range(len(candidatos)):
    print('-'*30,'\nVotos para', candidatos[cont], end=':')
    if votos.count==0:
        print('0')
        cont+=1
    else:
        print(votos.count(cont+1))
        cont+=1
porcentoNulo=(votos.count(5)/len(votos))*100
porcentoBranco=(votos.count(6)/len(votos))*100
print('\n% Nulos:',round(porcentoNulo,2), '%\n% Brancos:', round(porcentoBranco,2),'%')    
```

```python
'''Exerecício 45 - Desenvolver um programa para verificar a nota do aluno em uma prova com 10 questões, 
o programa deve perguntar ao aluno a resposta de cada questão e ao final comparar com o gabarito da prova 
e assim calcular o total de acertos e a nota (atribuir 1 ponto por resposta certa). 
Após cada aluno utilizar o sistema deve ser feita uma pergunta se outro aluno vai utilizar o sistema. 
Após todos os alunos terem respondido informar:
Maior e Menor Acerto;
Total de Alunos que utilizaram o sistema;
A Média das Notas da Turma.
Gabarito da Prova:

01 - A
02 - B
03 - C
04 - D
05 - E
06 - E
07 - D
08 - C
09 - B
10 - A
Após concluir isto você poderia incrementar o programa permitindo que o professor digite o gabarito da prova antes dos 
alunos usarem o programa.'''

gabarito = ['a','b','c','d','e','e','d','c','b','a']
certa=1
pontuacao=[]

perg=True
aluno=1
while perg!='N':
    resposta=[]
    print('\nAluno', aluno)
    for g in range(len(gabarito)):
        resp=str(input('Insira a letra da resposta correta:'))
        if resp == gabarito[g]:
            resposta.append(resp)
        else:
            print('[ERRADA]') 
    pont=len(resposta)
    pontuacao.append(pont)
    
    perg=str(input('Adicionar próximo aluno? [S/N]')).upper()
    if perg =='S':
        aluno+=1
    else:
        break
    
print('='*30)
print('Maior acerto:', max(pontuacao))
print('Menor acerto:', min(pontuacao))
print(len(pontuacao), 'alunos utilizaram o sistema')
media=sum(pontuacao)/len(pontuacao)
print('Média da turma: ', media)
print('Gabarito correto: ',gabarito)
```

```python
'''Exercício 46 - Em uma competição de salto em distância cada atleta tem direito a cinco saltos. 
No final da série de saltos de cada atleta, o melhor e o pior resultados são eliminados. 
O seu resultado fica sendo a média dos três valores restantes. Você deve fazer um programa que receba o nome e as 
cinco distâncias alcançadas pelo atleta em seus saltos e depois informe a média dos saltos conforme a descrição 
acima informada (retirar o melhor e o pior salto e depois calcular a média). 
Faça uso de uma lista para armazenar os saltos. Os saltos são informados na ordem da execução, 
portanto não são ordenados. O programa deve ser encerrado quando não for informado o nome do atleta.
A saída do programa deve ser conforme o exemplo abaixo:
Atleta: Rodrigo Curvêllo

Primeiro Salto: 6.5 m
Segundo Salto: 6.1 m
Terceiro Salto: 6.2 m
Quarto Salto: 5.4 m
Quinto Salto: 5.3 m

Melhor salto:  6.5 m
Pior salto: 5.3 m
Média dos demais saltos: 5.9 m

Resultado final:
Rodrigo Curvêllo: 5.9 m'''

saltos=[]
cont=1
nome=str(input('Atleta: '))
for pos in range(5):
    print(pos+1,'º salto: ')
    salto=float(input('Valor do salto: '))
    saltos.append(salto)

print('\nAtleta', nome)
print('-'*30)
for pos in range(5):
    print(pos+1,'º salto - ',saltos[pos],'m')
print('-'*30)
print('Melhor salto: ', max(saltos))
print('Pior salto: ', min(saltos))
for s in saltos:
    if s==max(saltos) and s==min(saltos):
        saltos.remove(s)
media=sum(saltos)/len(saltos)
print('Média dos demais saltos: ',media,'m')
print('-'*30)
print('RESULTADO FINAL')
print(nome,':', media,'m')
```

```python
'''Exercício 47 - Em uma competição de ginástica, cada atleta recebe votos de sete jurados. A melhor e a pior nota são 
eliminadas. A sua nota fica sendo a média dos votos restantes. Você deve fazer um programa que receba o 
nome do ginasta e as notas dos sete jurados alcançadas pelo atleta em sua apresentação e depois informe a sua média, 
conforme a descrição acima informada (retirar o melhor e o pior salto e depois calcular a média com as notas restantes).
As notas não são informados ordenadas. Um exemplo de saída do programa deve ser conforme o exemplo abaixo:
Atleta: Aparecido Parente
Nota: 9.9
Nota: 7.5
Nota: 9.5
Nota: 8.5
Nota: 9.0
Nota: 8.5
Nota: 9.7

Resultado final:
Atleta: Aparecido Parente
Melhor nota: 9.9
Pior nota: 7.5
Média: 9,04
'''
notas=[]
nome=str(input('Atleta: '))
for n in range(7):
    print(n+1, 'º Nota')
    nota=float(input('Nota: '))
    notas.append(nota)
    
print('\n Atlela: ',nome)
print('-'*30)
for n in range(7):
    print(n+1,'º Nota: ',notas[n])
print('\nRESULTADO FINAL')
print('Atleta: ',nome)
print('Melhor nota: ', max(notas))
print('Pior nota: ', min(notas))
for n in notas:
    if n==max(notas) and n==min(notas):
        n.remove(notas)
media=sum(notas)/len(notas)
print('Média: ',round(media, 2))
```

```python
'''Exercício 49 - Faça um programa que mostre os n termos da Série a seguir:
  S = 1/1 + 2/3 + 3/5 + 4/7 + 5/9 + ... + n/m. 
Imprima no final a soma da série.
'''

n=1
m=1
n_lista=[]
m_lista=[]
print('S =', end=' ')
while n <= 6-1:
    print(n,'/',m,' + ', end=' ')
    n_lista.append(n)
    m_lista.append(m)
    n+=1
    m+=2
print(n,'/',m,'=',sum(n_lista),'/',sum(m_lista))
```

```python
'''Exercício 50 - Sendo H= 1 + 1/2 + 1/3 + 1/4 + ... + 1/N, Faça um programa que calcule o valor de H com N termos.
'''
n=1
h=1
n_lista=[]
h_lista=[]
print('H = 1 +', end=' ')

while n <8:
    print('',h, '/', '+', end='')
    n_lista.append(n)
    h_lista.append(h)
    n+=1
print(h,'/','-',sum(h_lista), '/',sum(n_lista),'-',round(sum(h_lista)/sum(n_lista)))
    
```
